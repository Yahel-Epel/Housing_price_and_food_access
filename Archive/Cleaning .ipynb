```python
import pandas as pd
```

```python
# Files to load
zillow_zip_csv = "../Recourses/Zillow/Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv"

# Read the CSV into a DataFrame
zillow_zip_df = pd.read_csv(zillow_zip_csv)
zillow_zip_df.head(20)
```

```python
ca_zillow_zip_df = zillow_zip_df.loc[zillow_zip_df["State"] == "CA", :]
print(ca_zillow_zip_df.shape)
ca_zillow_zip_df.head(20)
```

```python
# Files to load
zip_tract_csv = "../Recourses/ZIP_TRACT_122019.csv"

# Read the CSV into a DataFrame
zip_tract_df = pd.read_csv(zip_tract_csv)
zip_tract_df.head(20)
```

```python
zip_tract_df =zip_tract_df[["ZIP", "TRACT"]]
zip_tract_df.head(20)
```

```python
zip_tract_df =zip_tract_df.rename(columns= {"TRACT": "CensusTract"})
zip_tract_df.head(20)
```

```python
# Files to load
food_access_csv = "../Recourses/Food_access/Food_access_2019.csv"

# Read the CSV into a DataFrame
food_access_df = pd.read_csv(food_access_csv, encoding='unicode_escape')
food_access_df.head(20)
```

```python
food_acces_ca_df= food_access_df.loc[food_access_df["State"] == "California", :]
print(food_acces_ca_df.shape)
food_acces_ca_df.head(20)
```

```python
tract_merge_df = pd.merge(food_acces_ca_df, zip_tract_df, on="CensusTract", how = "left")
print(tract_merge_df.shape)
tract_merge_df.head(5)
```
