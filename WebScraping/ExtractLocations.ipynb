```python
# Dependencies and Setup
import requests
import gmaps
import pandas as pd
import numpy as np
# Import API key
from config import g_key
```

```python
coordinates = [["32.8609,-117.2336"],["33.7222,-116.3745"],["33.6846,-117.8265"],["33.6846,-117.8265"],["34.1064,-117.5931"],
["33.7701,-118.1937"],["33.8358,-118.3406"],["34.1425,-118.2551"],["34.0736,-118.4004"],["34.1490, -118.4514"],["34.0195,-118.4912"],
["34.3917,-118.5426"],["34.1706,-118.8376"],["34.0919,-118.6021"],["36.9675,-121.9980"],["34.1975,-119.1771"],
["34.4208,-119.6982"],["36.7378,-119.7871"],["35.1428,-120.6413"],["39.0968,-120.0324"],["38.7521,-121.2880"],
["36.9741,-122.0308"],["37.3382,-121.886"],["37.5485,-121.9886"],["37.4419,-122.1430"],["37.7749,-122.4194"],
["37.8044,-122.2712"],["37.7644,-121.9540"],["37.9480,-122.0608"],["38.1074,-122.5697"],["38.2919,-122.4580"],["38.4404,-122.7141"]]

params = {
    "radius": 50000,
    "key": g_key,
    "name": "Whole Foods Market California",
}
lists = []
# Use base URL to search for whole foods
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

for coordinate in coordinates:

    params["location"]=coordinate
    # params{"location":[xxxx,xxxx]}
    
    # Make request and get the JSON data from the search.
    whole_foods = requests.get(base_url, params=params).json()
    lists.append(whole_foods)
    
    # base_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/output?"+parameter
print(base_url)



print(lists)
```

```python
WFLoc_df1 = pd.DataFrame()
WFLoc_df2 = pd.DataFrame()
WFLoc_df3 = pd.DataFrame()
WFLoc_df4 = pd.DataFrame()

for thislist in lists:
    
#     WFLoc_df = pd.DataFrame()
    
    for i in thislist['results']:
        
        name = i['name']
        address = i["vicinity"]
        lat = i["geometry"]["location"]["lat"]
        lng = i["geometry"]["location"]["lng"]
        WFLoc_df1 = WFLoc_df1.append([[name]])
        WFLoc_df2 = WFLoc_df2.append([[address]])
        WFLoc_df3 = WFLoc_df3.append([[lat]])
        WFLoc_df4 = WFLoc_df4.append([[lng]])
#         print(i)
#         WFLoc_df['Name'] = WFLoc_df.append(name)
        

        print(WFLoc_df2)
        
```

```python
WFLoc_df1 = WFLoc_df1.rename(columns={0:'store_name'})
WFLoc_df2 = WFLoc_df2.rename(columns={0:'address'})
WFLoc_df3 = WFLoc_df3.rename(columns={0:'lat'})
WFLoc_df4 = WFLoc_df4.rename(columns={0:'lng'})
WFLoc_df4
```

```python
dataframes = [WFLoc_df1,WFLoc_df2,WFLoc_df3,WFLoc_df4]
WFLoc_df = pd.concat(dataframes, axis = 1)
WFLoc_df
```

```python
WFLoc_df = WFLoc_df.drop_duplicates()
WFLoc_df
```

```python
WFLoc_df = WFLoc_df[WFLoc_df.store_name == "Whole Foods Market"]
WFLoc_df
```

```python
# Create the output File (CSV)
output_data_file = "WholeFoodsLocations.csv"
# Export the into a csv
WFLoc_df.to_csv(output_data_file, index_label="")
```
