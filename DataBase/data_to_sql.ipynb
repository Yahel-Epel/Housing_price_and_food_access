```python
import pandas as pd
import sqlite3
```

```python
housing_value_df = pd.read_csv("../Clean_Zillow_Data/housing_value_clean.csv",index_col=0)
housing_value_df.head()
```

```python
# Creates the SQLite Engine
conn = sqlite3.connect('housing_food_access_db.sqlite')
```

```python
housing_value_df.to_sql('housing_value', conn, if_exists='replace', index=False)
```

```python
pd.read_sql('select * from housing_value', conn)
```
