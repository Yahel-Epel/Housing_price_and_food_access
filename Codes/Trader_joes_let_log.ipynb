```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
df = pd.read_csv("../Food_access/store_info.csv")
df.head()
```

```python
df["full address"] = df["street"]+[", "]+df["city"]+[", "]+ df["state"]+[", "]+ df["zip"]
df.head()
```

```python
df = df.loc[df["state"] == "CA", :]
df.head()
```

```python
#!pip install geopy
```

```python
locator = Nominatim(user_agent="http")
```

```python
from geopy.extra.rate_limiter import RateLimiter
```

```python
# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
# 2- - create location column
df['location'] = df["full address"].apply(geocode)
```

```python
df.head()
```

```python
def my_split(val):
    if val:return list(val)[1]
    return None
    
    
```

```python
df['point'] = df['location'].apply(my_split)
```

```python
df.head()
```

```python
df['point2'] = df['point'].tolist()
```

```python
df['latitude'] = ""
df['longitude'] = ""
```

```python
for i, x in df.iterrows():
    print(i)
    if df.loc[i, 'point2']:
        df.loc[i, 'latitude'] = df.loc[i, 'point2'][0]
        df.loc[i, 'longitude'] = df.loc[i, 'point2'][1]
```

```python
df.head()
```

```python
df[['latitude', 'longitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)
```
