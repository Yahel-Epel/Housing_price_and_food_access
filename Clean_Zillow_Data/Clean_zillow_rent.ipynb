```python
import pandas as pd
```

```python
# Files to load
zillow_rent_csv = "../Recourses/Zillow/Zip_ZORI_AllHomesPlusMultifamily_Smoothed.csv"

# Read the CSV into a DataFrame
zillow_rent_df = pd.read_csv(zillow_rent_csv)
zillow_rent_df
```

```python
zillow_rent_df.dtypes
```

```python
for column in zillow_rent_df.columns:
    print(f"Column {column}: {zillow_rent_df[column].isnull().sum()} null values and type: {type(column)}")
```

```python
clean_zillow_rent_df= zillow_rent_df[["RegionID","RegionName", "MsaName", "2015-12", "2019-12"]]
clean_zillow_rent_df
```

```python
# new data frame with split value columns
new_zillow_rent_df = clean_zillow_rent_df["MsaName"].str.split(", *", n = 1, expand = True)
  
# making separate first name column from new data frame
clean_zillow_rent_df["City"]= new_zillow_rent_df[0]
  
# making separate last name column from new data frame
clean_zillow_rent_df["State"]= new_zillow_rent_df[1]
   
# df display
clean_zillow_rent_df
```

```python
ca_zillow_rent_df = clean_zillow_rent_df.loc[clean_zillow_rent_df["State"] == "CA", :]
ca_zillow_rent_df
```

```python
ca_zillow_rent_df = ca_zillow_rent_df.rename(columns={"RegionName": "ZIP"})
ca_zillow_rent_df
```

```python
for column in ca_zillow_rent_df.columns:
    print(f"Column {column}: {ca_zillow_rent_df[column].isnull().sum()} null values and type: {type(column)}")
```

```python
# Files to load
zip_tract_csv = "../Recourses/ZIP_TRACT_122019.csv"

# Read the CSV into a DataFrame
zip_tract_df = pd.read_csv(zip_tract_csv)
zip_tract_df
```

```python
# Merge the two DataFrames together based on the Dates they share
ca_rent_df = pd.merge(ca_zillow_rent_df, zip_tract_df, on="ZIP")
# Merge the two DataFrames together based on the Dates they share
ca_rent_df
```

```python
ca_rent_df= ca_rent_df[["ZIP","TRACT", "State", "City", "2015-12", "2019-12"]]
ca_rent_df
```

```python
ca_rent_df.count()
```

```python
no_null_ca_rent_df = ca_rent_df.dropna(how='any')
no_null_ca_rent_df.count()
```

```python
ca_rent_clean_df =  no_null_ca_rent_df.rename(columns={"TRACT": "CensusTract"})
ca_rent_clean_df
```

```python
ca_rent_clean_df.to_csv("clean_ca_rent.csv")
```
